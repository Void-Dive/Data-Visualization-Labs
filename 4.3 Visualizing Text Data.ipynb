```python
plt.figure(figsize=(8,5))
plt.scatter(df["Total_Vote"], df["Helpful_Vote"], color='blue', alpha=0.5, s=20)

plt.title("Helpful Votes vs Total Votes", fontsize=14)
plt.xlabel("Total Votes")
plt.ylabel("Helpful Votes")
plt.grid(alpha=0.3)

plt.show()
```

### Helpful Votes vs Total Votes

This scatterplot shows the relationship between total votes and helpful votes for Spider-Man movie reviews. There is a strong positive correlation, indicating that reviews with more total votes also tend to receive more helpful votes. This suggests that highly engaged reviews are more likely to be considered useful by other users.
