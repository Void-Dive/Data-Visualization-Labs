```python
import pandas as pd

df = pd.read_csv('Employee_productivity.csv')

df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def calculate_r2(x, y):
    correlation = np.corrcoef(x, y)[0, 1]
    return correlation ** 2
```

```python
x = df['YearsAtCompany']
y = df['ProductivityScore']

plt.scatter(x, y)
plt.title('Years at Company vs Productivity')
plt.xlabel('Years at Company')
plt.ylabel('Productivity')
plt.grid(True)
plt.show()

r2_years = calculate_r2(x, y)
print("R² (Years):", r2_years)
```

### Years at Company vs Productivity

There is no clear relationship between years at the company and productivity. The R² value is extremely low, indicating that experience does not meaningfully predict productivity.

```python
x = df['HoursWorked']
y = df['ProductivityScore']

plt.scatter(x, y)
plt.title('Hours Worked vs Productivity')
plt.xlabel('Hours Worked')
plt.ylabel('Productivity')
plt.grid(True)
plt.show()

r2_hours = calculate_r2(x, y)
print("R² (Hours):", r2_hours)
```

### Hours Worked vs Productivity

There is no strong relationship between hours worked and productivity. The R² value is very low, suggesting that working more hours does not significantly increase productivity.

```python
x = df['FocusScore']
y = df['ProductivityScore']

plt.scatter(x, y)
plt.title('Focus Score vs Productivity')
plt.xlabel('Focus Score')
plt.ylabel('Productivity')
plt.grid(True)
plt.show()

r2_focus = calculate_r2(x, y)
print("R² (Focus):", r2_focus)
```

### Focus Score vs Productivity

There is a strong positive relationship between focus score and productivity. The high R² value shows that employees with higher focus tend to have significantly higher productivity.

### Core Feature Comparison

Among the three features, focus score has the strongest relationship with productivity based on its high R² value. Years at the company and hours worked show almost no relationship, making them poor predictors of productivity.

```python
x = df['TeamMeetingsPerWeek']
y = df['ProductivityScore']

plt.scatter(x, y)
plt.title('Meetings vs Productivity')
plt.xlabel('Meetings per Week')
plt.ylabel('Productivity')
plt.grid(True)
plt.show()

r2_meetings = calculate_r2(x, y)
print("R² (Meetings):", r2_meetings)
```

### Meetings vs Productivity

There is a very weak relationship between meetings per week and productivity. The R² value is low, indicating that meeting frequency does not strongly impact productivity.

### Model Evaluation

Based on the R² values, focus score is the strongest predictor of productivity. Hours worked and years at the company show almost no relationship with productivity, while meetings per week shows only a very weak relationship.

This suggests that productivity is primarily driven by how focused employees are, rather than how long they work, how experienced they are, or how many meetings they attend.

### Final Summary and Recommendations

Based on the analysis, focus score has the strongest relationship with productivity, indicating that employees who are more focused tend to perform significantly better. In contrast, years at the company and hours worked show little to no relationship with productivity, suggesting that experience and longer working hours do not necessarily lead to better performance. Meetings per week also show a very weak relationship with productivity.

These results suggest that the company should prioritize improving employee focus rather than increasing hours or relying on experience alone. Reducing unnecessary meetings and minimizing workplace distractions may help employees maintain higher levels of focus. Additionally, improving work environments, especially for in-person employees, could further enhance productivity.
