```python
import plotly.express as px

df = px.data.gapminder()
df.head()
```

### Dataset Description

This notebook uses the Gapminder dataset built into Plotly Express. The dataset contains information for countries across multiple years, including country name, continent, year, life expectancy, population, GDP per capita, and country codes. It is appropriate for this lab because it includes both spatial data and time-series data, allowing trends to be analyzed across soace and time.

```python
fig = px.choropleth(
    df,
    locations="iso_alpha",
    color="gdpPercap",
    hover_name="country",
    animation_frame="year",
    color_continuous_scale="Viridis",
    title="GDP Per Capita Over Time"
)

fig.show()
```

### GDP Per Capita Over Time

This visualization shows how GDP per capita changes across countries over time. Overall, GDP increases globally from 1952 to 2007, but the growth is uneven. Countries like the United States, Canada, and parts of Europe show consistently higher GDP levels, while many regions in Africa and parts of Asia remain lower throughout the timeline. This highlights the economic gap between developed and developing regions.

```python
fig = px.choropleth(
    df,
    locations="iso_alpha",
    color="lifeExp",
    hover_name="country",
    animation_frame="year",
    color_continuous_scale="Plasma",
    title="Life Expectancy Over Time"
)

fig.show()
```

### Life Expectancy Over Time

This visualization illustrates changes in life expectancy across the world. Over time, life expectancy increases in nearly all regions, with noticeable improvements from 1952 to 2007. Developed regions such as North America, Europe, and Australia consistently have higher life expectancy, while some regions in Africa lag behind. However, the overall upward trend suggests global improvements in healthcare and living conditions.

```python
fig = px.choropleth(
    df,
    locations="iso_alpha",
    color="pop",
    hover_name="country",
    animation_frame="year",
    color_continuous_scale="Inferno",
    title="Population Growth Over Time"
)

fig.show()
```

### Population Growth Over Time

This visualization shows population growth across countries over time. Population increases significantly from 1952 to 2007, with countries like China and India showing the most dramatic growth. Other regions grow more slowly in comparison. The visualization highlights how population distribution becomes more concentrated in certain areas over time.
