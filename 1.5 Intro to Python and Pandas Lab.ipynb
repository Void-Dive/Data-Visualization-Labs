```python
# 1. What are the top 10 highest-rated anime in the dataset?

df['Score'] = pd.to_numeric(df['Score'], errors='coerce')
df.sort_values(by='Score', ascending=False)[['Name', 'Score']].head(10)
```

The dataset initially contained non-numeric values such as "UNKNOWN" in the score column. After cleaning the data by converting the Score column to numeric values, the top 10 highest-rated anime were identified correctly. These anime have the highest user ratings in the dataset, showing which shows are most highly regarded by viewers.

```python
# 2. What are the most common genres in the dataset?

df['Genres'] = df['Genres'].replace('UNKNOWN', None)
df['Genres'].str.split(', ').explode().value_counts().head(10)
```

The most common genres in the dataset were identified by splitting and counting each genre individually. Before analysis, missing or "UNKNOWN" values were removed to ensure more accurate results. The most frequent genres include Action, Comedy, and Fantasy, showing that these types of anime are the most common in the dataset.

```python
# 3. What is the average score of anime by type (TV, Movie, etc.)?

df.groupby('Type')['Score'].mean().sort_values(ascending=False)
```

The average score of anime by type was calculated by grouping the dataset based on the Type column and then finding the mean score for each group. This shows how different formats, such as TV series and movies, compare in terms of average user ratings.

```python
# 4. How does the number of episodes relate to the score of an anime?

df['Episodes'] = pd.to_numeric(df['Episodes'], errors='coerce')
df[['Episodes', 'Score']].corr()
```

The relationship between the number of episodes and anime scores was analyzed using correlation. The result was approximately 0.08, which indicates a very weak positive relationship. This means that the number of episodes has little to no impact on the rating of an anime. Both short and long series can receive high or low scores independently.

```python
# 5. Which studios produce the highest-rated anime on average?

df.groupby('Studios')['Score'].mean().sort_values(ascending=False).head(10)
```

The average score by studio was calculated by grouping the dataset based on the studio and computing the mean score for each group. The results show which studios consistently produce higher-rated anime. This helps identify studios known for creating high-quality and well-received content.
